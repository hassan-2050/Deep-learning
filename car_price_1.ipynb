```python
import pandas as pd
```

```python
d=pd.read_csv("/content/drive/MyDrive/data_sets/CarPrice_Assignment.csv")
d
```

```python
data_string=d.select_dtypes(include="object")  #we are only taking string colomns i.e object columns
data_number=d.select_dtypes(exclude="object")# we are choosing only numerical columns
data_number
```

```python
data_target=data_number.loc[:,"price",] #price coloumn ko ilag kar lia
data_target=data_target.values
data_target
car_id=data_number.loc[:,"car_ID"]
```

```python
data_string
```

```python
data_number=data_number.drop(columns=["price","car_ID"])# data frame main se price column ko nikal diya kyun ke price ko predict karna hai aur car id ka faida ni 
data_string=data_string.drop(columns="CarName")#car name ka prediction main koi role ni

```

#NORMALIZATION

```python
mean=data_number.mean(axis=0)
data_number-=mean
std=data_number.std(axis=0)
data_number/=std
```

#dummies

```python
data_string=pd.get_dummies(data_string)
```

```python
data_string
```

```python
data_frame=pd.concat([data_number,data_string],axis=1)
data_frame
```

#seprating test val train and respective labels

```python
import numpy as np

train=data_frame.loc[:103]
train_l=data_frame.loc[:103]
test=data_frame.loc[103:164]
test_l=data_frame.loc[103:164]
val=data_frame.loc[164:]
val_l=data_frame.loc[164:]
```

```python
from keras import layers
from keras import models
model=models.Sequential()
model.add(layers.Dense(10,activation="relu",input_shape=(52,)))
model.add(layers.Dense(8,activation="relu"))
model.add(layers.Dense(6,activation="relu"))
model.add(layers.Dense(1))




```

```python
model.compile(optimizer="rmsprop",loss="mse",metrics=["mae"])
```

```python
model.fit(train,train_l,epochs=300,batch_size=100,validation_data=(val,val_l)
    


)
```

```python
model1=models.Sequential()
model1.add(layers.Dense(10,activation="relu",input_shape=(52,)))
model1.add(layers.Dense(8,activation="relu"))
model1.add(layers.Dense(6,activation="relu"))
model1.add(layers.Dense(1))
```

```python
y_pred = model.predict(test)
y_pred
```

```python
y_pred=np.array(y_pred).reshape(62,)
```
